```python
# Imports
# import matplotlib
# matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
import matplotlib.patches as patches

from layout import *
from mathutils import *
```

```python
# Placeholder for a single image
class ImagePlaceholder(Rectangle):
    def __init__(self):
        Rectangle.__init__(self)
        self.text = None

    @staticmethod
    def create(w, h):
        result = ImagePlaceholder()
        result.text = str(w) + "x" + str(h)
        result.width = w
        result.height = h
        return result
```

```python
# Helper function, which visualises a layout
def visualise_layout(ax, images, canvas, bg_rects=[]):    
    #ax.set_ylim(canvas.height + 10, -10)
    #ax.set_xlim(-10, canvas.width + 10)
    ax.set_ylim(canvas.height, 0)
    ax.set_xlim(0, canvas.width)
    ax.set_aspect('equal')
    #ax.axis('off')
#     ax.set_title('layout')

    ax.add_patch(patches.Rectangle((0, 0), canvas.width, canvas.height, linewidth=1,
                              alpha=0.7, linestyle="solid", facecolor='none', edgecolor="#000000"))

    for bg_rect in bg_rects:
        #print("Drawing BG-" + str(bg_rect))
        p = patches.Rectangle((bg_rect.x, bg_rect.y), bg_rect.width, bg_rect.height, linewidth=1,
                              alpha=0.6, linestyle="solid", facecolor='#33FF33', edgecolor="#00CC00", hatch='x')
        ax.add_patch(p)

    for img in images:
        #print("Drawing", str(img))
        p = patches.Rectangle((img.x, img.y), img.width, img.height, linewidth=1,
                              alpha=0.7, linestyle="solid", facecolor='none', edgecolor="#FF6B6B", hatch='/')
        ax.add_patch(p)
    
    return ax

# Given a layout generator functor, generate and plot 4 layouts (as provided by fg_rects, bg_rects, canvas)
def visualise_layout_generator(generator, fg_rects, bg_rects, canvas):
    fig = plt.figure(dpi=144)
    for i in range(1,5):
        ax = plt.subplot(2,2,i)
        generate_layout_1(fg_rects, bg_rects, canvas)
        visualise_layout(ax, fg_rects, canvas, bg_rects)
    fig.show()
    plt.show()
```

```python
def pertubate_generator(generator, image_width, image_height, max_fg=5, max_bg=5, canvas=Rectangle(0, 0, 1920, 1080)):
    for num_fg in range(1, max_fg+1):
        for num_bg in range(1, max_bg+1):
            fg_rects = [ImagePlaceholder.create(image_width, image_height) for i in range(num_fg)]
            bg_rects = [ImagePlaceholder.create(image_width, image_height) for i in range(num_bg)]
            print("Plotting {} foreground and {} background images".format(num_fg, num_bg))
            visualise_layout_generator(generate_layout_1, fg_rects, bg_rects, canvas)
```
